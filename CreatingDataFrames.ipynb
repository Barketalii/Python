```python
Employee={
    'Name':['Ahmad','Sana','Bilal','Mehak'],
    'Department':['HR','IT','Finance','IT'],
    'Salary':[50000,60000,55000,65000]}

Employee
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(Employee)
```

```python
df
```

```python
df['Salary']=(df['Updated_Salary']*1.10).astype(int)
```

```python
df
#df['Updated_Salary'] = (df.pop('Salary') * 1.10).round(0).astype(int) this method is used to update the column by the number u want and also replace
#the column with new name
```

```python
Updated_Salary,Salary=df[['Updated_Salary','Salary']].to_numpy().T
```

```python
Updated_Salary,Salary
```

```python
df.to_csv('Employee.csv',index=False)
```
